## Analysis Review

##### Dimensionality Reduction (PCA / UMAP)
> ##### Treatment-seeking behavior is driven by a latent factor (PC4) capturing salience + readiness to act, influenced by
> > ##### Personal experience (family history)
> > ##### Perceived impact (work interference)
> > ##### Observed outcomes (obs consequence)
> > ##### Willingness to take leave (leave)
> ##### Not everyone is equally primed to seek treatment
##### Clustering / Segmentation
> ##### Clustering revealed distinct groupings
> > ##### Cluster 0: Lower support, moderate readiness
> > ##### Cluster 1: High readiness, but uninformed / uncertain environment
> > ##### Cluster 2: Moderate support, clearer pathways
> ##### Even with similar needs, people operate in different environments
##### Propensity Modeling
> ##### PC4 was the strongest predictor of seeking treatment across models
> > ##### Within decision trees, a high PC4 (>0.5) nearly always indicated treatment sought
> ##### When PC4 was low, environmental factors determined treatment-seeking behavior
##### 2x2 Framework
> ##### Treatment-seeking is not just a function of readiness; it depends on the interaction between readiness and environment
> > ##### High readiness: people seek treatment regardless
> > ##### Low readiness: environment becomes decisive

#### While individual mental health salience is the primary driver of treatment-seeking, workplace environment plays a critical enabling role, particularly for individuals who are not inherently inclined to seek care

## Recommendation for Organizations

##### 1. Increase visibility and clarity of resources
> ##### Cluster 1 (high need, low support) was identified as the most vulnerable segment
> > ##### Responses were largely 'I don't know' for various workplace support features
> > ##### Showed the highest proportion of High_PC4 respondents and treatment within that population
> > ##### But tied for lowest treatment rate among Low_PC4 respondents
##### Ensure high-need individuals can successfully translate intent into action
> ##### Centralized, easy-to-understand resource hubs
> ##### Clear 'how to get help' pathways
> ##### Manager training to direct employees to resources
> ##### Proactive outreach, not just passive availability

##### 2. Reduce friction to increase uptake
> ##### Cluster 2-type profiles provide the biggest opportunity: low readiness & high support
> > ##### They can be influenced
> > ##### They respond to environment
##### Target 'moveable' employees by making treatment easier to initiate
> ##### Normalize help-seeking
> ##### Simplify access
> ##### Improve behavioral nudges

##### 'Drive Impact'
> ##### Cluster 2 presents the largest opportunity to increase overall treatment uptake
##### 'Prevent Failure'
> ##### Cluster 1 represents the highest-risk population, where failure to access care may have the most severe consequences

#### Interventions should prioritize both efficiency (maximizing uptake) and equity (ensuring access for vulnerable groups)
